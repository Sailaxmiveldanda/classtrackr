```python
import pandas as pd
import os
```

```python
file_path = '/content/drive/MyDrive/Datasets/Spring2023Outcome.csv'
original_df = pd.read_csv(file_path)
required_columns = [ 'account id', 'account name', 'course id', 'course name', 'course sis id', 'assessment id', 'assessment title', 'learning outcome id', 'learning outcome name', 'learning outcome rating']

if all(column in original_df.columns for column in required_columns):
    selected_df = original_df[required_columns]
    selected_df = selected_df.dropna()
    grouped_df = selected_df.groupby(['account name', 'account id', 'course id', 'course name', 'course sis id', 'assessment id', 'assessment title', 'learning outcome id', 'learning outcome name', 'learning outcome rating']).apply(lambda group: group.reset_index(drop=True))
    file_name = os.path.basename(file_path)
    output_file_name = file_name.replace('.csv', '_cleaned_grouped.csv')
    grouped_df.to_csv(output_file_name, index=False)
    print(f"Grouped data saved to '{output_file_name}'")
else:
    print("Not all required columns are present in the original data.")
```
